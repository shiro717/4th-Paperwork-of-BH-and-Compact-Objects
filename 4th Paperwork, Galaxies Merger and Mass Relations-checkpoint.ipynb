```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

num_galaxies = 10001
num_mergers = 200
min_galaxy_mass = 1e4
max_galaxy_mass = 1e7
min_bh_mass = 1e1
max_bh_mass = 1e4
num_trials = 200

def galaxy_merger_simulation():
    galaxy_masses = np.random.lognormal(mean=np.log((min_galaxy_mass + max_galaxy_mass) / 2), 
                                         sigma=1, size=num_galaxies)
    bh_masses = np.random.lognormal(mean=np.log((min_bh_mass + max_bh_mass) / 2), 
                                     sigma=1, size=num_galaxies)

    galaxy_masses_history = [galaxy_masses.copy()]
    bh_masses_history = [bh_masses.copy()]

    for merger_step in range(num_mergers):
        num_to_select = min(100, len(galaxy_masses) // 2 * 2)
        selected_indices = np.random.choice(len(galaxy_masses), size=num_to_select, replace=False)

        new_galaxies = []
        new_bh_masses = []
        for i in range(0, num_to_select, 2):
            galaxy1_idx = selected_indices[i]
            galaxy2_idx = selected_indices[i + 1]

            new_galaxy_mass = galaxy_masses[galaxy1_idx] + galaxy_masses[galaxy2_idx]
            new_bh_mass = bh_masses[galaxy1_idx] + bh_masses[galaxy2_idx]

            new_galaxies.append(new_galaxy_mass)
            new_bh_masses.append(new_bh_mass)

        remaining_indices = list(set(range(len(galaxy_masses))) - set(selected_indices))
        galaxy_masses = np.concatenate([galaxy_masses[remaining_indices], new_galaxies])
        bh_masses = np.concatenate([bh_masses[remaining_indices], new_bh_masses])

        galaxy_masses_history.append(galaxy_masses.copy())
        bh_masses_history.append(bh_masses.copy())

    return galaxy_masses, bh_masses, galaxy_masses_history, bh_masses_history

final_galaxy_masses_all = []
final_bh_masses_all = []
galaxy_masses_history_all = []
bh_masses_history_all = []

for _ in range(num_trials):
    final_galaxy_masses, final_bh_masses, galaxy_masses_history, bh_masses_history = galaxy_merger_simulation()
    final_galaxy_masses_all.append(final_galaxy_masses)
    final_bh_masses_all.append(final_bh_masses)
    galaxy_masses_history_all.append(galaxy_masses_history)
    bh_masses_history_all.append(bh_masses_history)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
for i in range(num_trials):
    plt.scatter(final_galaxy_masses_all[i], final_bh_masses_all[i], alpha=0.8, s=10)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Galaxy Mass (Solar Mass)')
plt.ylabel('Black Hole Mass (Solar Mass)')
plt.title('Galaxy Mass versus Black Hole Mass (All Trials)')

initial_galaxy_masses = galaxy_masses_history_all[0][0]
initial_bh_masses = bh_masses_history_all[0][0]

plt.figure(figsize=(12, 8))

steps = [50, 100, 150, 200]
for step in steps:
    plt.subplot(2, 2, steps.index(step) + 1)
    plt.scatter(galaxy_masses_history_all[0][step], bh_masses_history_all[0][step], alpha=0.8, s=10)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Galaxy Mass (Solar Mass)')
    plt.ylabel('Black Hole Mass (Solar Mass)')
    plt.title(f'Galaxy Mass versus Black Hole Mass (Step {step})')

plt.tight_layout()
plt.show()
```
